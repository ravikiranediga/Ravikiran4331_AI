```python
import pandas as pd
import matplotlib.pyplot as plt


#step 1
data=pd.DataFrame({
    'Age':[22,25,30,35,40,45,50,55,60,65],
    'Income':[25000,27000,30000,35000,40000,45000,48000,55000,60000,80000]
})

#stwp 2
x_column="Age"
y_column="Income"

#step 3
data.plot.scatter(x=x_column,y=y_column)

#step 4
plt.title("Age vs. income sactter plot")
plt.xlabel('Age')
plt.ylabel('Income')


#step 5
plt.show()
```
