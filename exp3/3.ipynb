```python
print("Roll No:23G01A4361")
df=sns.load_dataset("titanic")
```

```python
print("Roll No:23G01A4361")
df.head()
```

```python

df=df[["survived","pclass","sex","age","fare","embarked","sibsp","parch"]]
```

```python
print("Roll No:23G01A4361")
df['sex']=df['sex'].map({'male':0,'female':1})
```

```python
print("Roll No:23G01A4361")
df['embarked']=df['embarked'].map({'C':0,'Q':1,'S':2})
```

```python
print("Roll No:23G01A4361")
df.head()
```

```python
print("Roll No:23G01A4361")
print(df.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='median')
```

```python
df['age']=imputer.fit_transform(df[['age']])
```

```python
print("Roll No:23G01A4361")
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
age_discretizer=KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='uniform')
```

```python
df['age_bins'] = age_discretizer.fit_transform(df[['age']])
```

```python
print("Roll No:23G01A4361")
age_discretizer
```

```python
fare_discretizer=KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='quantile')
```

```python
df['fare_bins']=fare_discretizer.fit_transform(df[['fare']])
```

```python
print("Roll No:23G01A4361")
print(df[['age','age_bins','fare','fare_bins']].head())
```

```python
from scipy import stats
```

```python
z_scores=np.abs(stats.zscore(df[['age','fare']]))
```

```python
print("Roll No:23G01A4361")
z_scores
```

```python
df_clean=df[(z_scores<3).all(axis=1)]
```
