```python
#importing libraries
import pandas as pd
import numpy as np

#Creating Sample Datasets
da={
    'X':[1,2,3,np.nan,5],
    'Y':[5,6,4,7,8],
    'Z':[np.nan,2,3,4,np.nan]
}
df=pd.DataFrame(da)
print("Original DataFrame")
display(df)

df_dropna_rows=df.dropna()
print("/n DataFrame after dropping rows with any Nan Values:")
display(df_dropna_rows)


df_dropna_columns=df.dropna(axis=1)
print("/n DataFrame after dropping rows with any Nan Values:")
display(df_dropna_columns)


df_dropna_specific=df.dropna(subset=['X'])
print("/n DataFrame after dropping rows with any Nan Value in columnss 'X':")
display(df_dropna_specific)
```

```python

import pandas as pd

data={
    'Product':['X','X','Y','Y','Z','Z','X','Y','Z'],
    'Region':['East','West','East','West','East','West','East','West','East'],
    'Sales':['Sim','Jack','Sim','Jack','Sim','Jack','Sim','Jack','Sim'],
    'Units':[12,15,28,31,19,13,22,33,22],
    'Profit':[299,400,289,399,488,299,488,288,374]
}
df=pd.DataFrame(data)
print("Original DataFrame:")
display(df)

pivot_table_units =pd.pivot_table(df,values="Units",index='Product',columns="Region",aggfunc="sum")
print("\n Pivot Table-Total Units by Product and Regions:")
display(pivot_table_units)


pivot_table_units =pd.pivot_table(df,values="Units",index='Product',columns="Sales",aggfunc="mean")
print("\n Pivot Table-Average profit per slaes for each product:")
display(pivot_table_units)


pivot_table_units =pd.pivot_table(df,values="Units",index='Product',columns="Region",aggfunc=['sum','mean'])
print("\n Pivot Table-Total and Average Units by Product and Regions:")
display(pivot_table_units)
```
