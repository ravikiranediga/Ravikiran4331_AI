```python
import pandas as pd
#Sample data frame
data=pd.DataFrame({
    'Product':['A','B','C'],
    'Q1_Sales':[150,200,250],
    'Q2_Sales':[100,220,270],
    'Q3_Sales':[170,210,260],
    'Q4_Sales':[160,200,250]
})

print("Original DataFrame(Wide Format):\n",data)


#Melt the dataframe to convert it to long format
melted_data=data.melt(id_vars='Product',
                      value_vars=['Q1_Sales','Q2_Sales','Q3_Sales','Q4_Sales'],
                      var_name='Quarter',
                      value_name='Sales')

print("\nMelted DataFrame (Long Format):\n",melted_data)
                      
```
