```python
import pandas as pd
```

```python
labels=['a','b','c','d','e']
data=[10,20,30,40,50]
Series_with_labels=pd.Series(data,index=labels)
print("Roll No:23G01A4331")
print("Pandas Series with Labels:")
print(Series_with_labels)
type(Series_with_labels)
```

```python
data_dict={"a":100,"b":200,"c":300,"d":400,"e":500}
Series_from_dict=pd.Series(data_dict)
print("Roll No:23G01A4331")
print("\nPandas Series from Dictionary:")
print(Series_from_dict)
```

```python
data_dict_df={
   "Name":["Ravi","Reenu","Bubu","Dudu","Lassie","Sravani"],
   "Age":[31,61,18,12,10,5],
   "Salary":[50000,60000,70000,80000,55000,45000]
}
print("Roll No:23G01A4331")
df= pd.DataFrame(data_dict_df)
print("\nPandas DataFrame:")
print(df)
type(data_dict_df)
```
